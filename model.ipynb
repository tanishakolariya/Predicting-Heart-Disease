```python
df_test  = pd.read_csv("test (1).csv")
df_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay
X_train = df.drop(columns=["id", "Heart Disease", "Heart Disease 2", "BP", "Cholesterol", "FBS over 120"])
y_train = df["Heart Disease 2"]
x_test = df_test.drop(columns=["id", "BP", "Cholesterol", "FBS over 120"])
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression())
])
pipeline.fit(X_train, y_train)
test_probs = pipeline.predict_proba(x_test)[:, 1]
print("test_probabilities:", test_probs)
scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='roc_auc')
print("CV ROC-AUC:", scores.mean())
scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='accuracy')
print("CV Accuracy:", scores.mean())
scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='precision')
print("CV Precision:", scores.mean())   
scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='recall')
print("CV Recall:", scores.mean())
```

```python
ConfusionMatrixDisplay.from_estimator(pipeline, X_train, y_train)
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
import pandas as pd

selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(X_train, y_train)
```

```python
feature_scores = pd.DataFrame({
    'Feature': X_train.columns,
    'Score': selector.scores_
})

print(feature_scores)
```

```python
feature_scores = feature_scores.sort_values(by='Score', ascending=False)
print(feature_scores)
```
